```python
# Make sure Product ID column is sorted
df = df.sort_values("Product_ID", ascending =True)
df
```

```python
def binary_search(df, column, target):
    """
    Perform binary search on a sorted DataFrame column.
    :param df: pandas DataFrame
    :param column: column name to search (must be sorted)
    :param target: value to search for
    :return: row index if found, else -1
    """
    left, right = 0, len(df) - 1
    
    while left <= right:
        mid = (left + right) // 2
        value = df.iloc[mid][column]
        
        if value == target:
            return mid   # Found row index
        elif value < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1   # Not found




search_id ='PID-10000022'	    # example product ID
result = binary_search(df, "Product_ID", search_id)

if result != -1:
    print("Product found:")
    print(df.iloc[result])
else:
    print(f"Product ID {search_id} not found")
```

```python
def sequential_search(df, column, target):
    """
    Perform sequential search on a DataFrame column.
    :param df: pandas DataFrame
    :param column: column name to search
    :param target: value to search for
    :return: row index if found, else -1
    """
    for i in range(len(df)):
        if df.iloc[i][column] == target:
            return i
    return -1


    
search_id ='PID-10000010'	    # example product ID
result = sequential_search(df, "Product_ID", search_id)

if result != -1:
    print("Product found:")
    print(df.iloc[result])
else:
    print(f"Product ID {search_id} not found")
```
